```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import joblib
```

```python
# Load the dataset
df = pd.read_csv('/content/drive/MyDrive/almabetter 1/Tast3/Churn_Modelling.csv')
df.head()
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
df.head()
```

```python
x = df.drop('Exited',axis=1)
y = df['Exited']
```

```python
# Train-test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)
```

```python
lr = LogisticRegression(max_iter=1000)
lr.fit(x_train,y_train)
```

```python
# Train Random Forest model
forest_model = RandomForestClassifier(n_estimators=100, random_state=42)
forest_model.fit(x_train, y_train)
```

```python
# Train Gradient Boosting model
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
xgb_model.fit(x_train, y_train)
```

```python
# Evaluate models
def evaluate_model(model, x_test, y_test):
    y_pred = model.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f'Accuracy: {accuracy:.2f}')
    print(classification_report(y_test, y_pred))
    print(confusion_matrix(y_test, y_pred))
```

```python
print("Logistic Regression:")
print(evaluate_model(lr, x_test, y_test))
```

```python
print("Random Forest")
print(evaluate_model(forest_model, x_test, y_test))
```

```python
print("Gradient Boosting ")
print(evaluate_model(xgb_model, x_test, y_test))
```

```python
# Save models
joblib.dump(lr, 'logistic_model.pkl')
joblib.dump(forest_model, 'forest_model.pkl')
joblib.dump(xgb_model, 'xgb_model.pkl')
```
